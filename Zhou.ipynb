# Analysis
After analyzing the data, I only need the first two columns. I also realized that some of the first columns have multiple neighborhoods within the columns. So I researched the neighborhoods that are in the same cell, and I found out that they are grouped together because they are right next to each other. So the logic I used is to make my dataset smaller why sorting out the top 10 neighborhoods with the highest number of gifted enrollment. After having a sorted dataset, there will be a graph to show the results. 

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("neighborhood_iep.csv")

d = df.iloc[:, :2].copy()
d.columns = ["neighborhood", "iep"]

d["iep"] = pd.to_numeric(d["iep"], errors="coerce").fillna(0)

top10 = d.sort_values("iep", ascending=False).head(10)

# My Graph
plt.figure()
bars = plt.bar(top10["neighborhood"], top10["iep"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("IEP enrollment")
plt.title("Top 10 Neighborhoods by IEP Enrollment")

plt.ylim(0, top10["iep"].max()*1.10 if top10["iep"].max() > 0 else 1)
ax = plt.gca()
for b in bars:
    y = b.get_height()
    ax.text(b.get_x()+b.get_width()/2, y, f"{int(y):,}", ha="center", va="bottom", fontsize=9)

plt.tight_layout()
plt.show()
```
